```python
import requests
import json
"""
The 'students' variable is already loaded from the URL in a previous cell.
Therefore, we can directly use the 'students' variable.
"""
data = students
print("JSON file loaded")
ref = db.reference("students")
for student in data["students"]:
  ref.child(str(student["id"])).set(student)
print("Data uploaded successfully!")
```

```python
students_ref = db.reference("students")
students_data = students_ref.get()
print(students_data)
```

```python
for student_id, info in students_data.items():
  print(info["name"])
```

```python
for student_id, info in students_data.items():
  print(info["name"], info["grades"])
```

```python
students_data = ref.get()
for key, student in students_data.items():
  name = student["name"]
  grades = student["grades"].values()
  avg = sum(grades) / len(grades)
  print(f"Student: {name}")
  print(f"Grades: {student['grades']}")
  print(f"Average: {avg:.2f}\n")
```
